¿Cuál es el porcentaje de valores nulos en cada columna? Presenta los resultados en un DataFrame, donde las columnas sean el nombre de las variables y el porcentaje de nulos.


```python
cuenta = df.isnull().sum()
porcentaje = (cuenta / df.shape[0] * 100).round(2)
df_porcentaje = porcentaje.reset_index()

new_cols = {"index": "variable", 0 : "porcentaje"}
df_porcentaje.rename(columns = new_cols, inplace = True)

df_porcentaje
```

Cread dos DataFrames nuevos, uno que incluya solo las variables categóricas y otro que incluya solo las variables numéricas.


```python
#Variables categóricas

df_categoricas = df.select_dtypes(include=['object','category']) 
df_categoricas.head()
```

Extraed los valores únicos su frecuencia. 

```python
for col in df.columns:
    print("------------------------------------------------------------", "\n")
    print(f"En la columna {col} tenemos los siguientes valores unicos")
    print(df[col].unique())
    print("\n", "------------------------------------------------------------", "\n")
```

```python
lista = []
for col in df.columns:
    df_col = df[col].value_counts()
    print("esta es una vuelta al for")
    lista.append(df_col)
```

```python
lista = []
for col in df.columns:
    df_col = pd.DataFrame(df[col].value_counts())
    dic = {col : df_col}
    lista.append(dic)
```

BONUS presenta los resultados en un DataFrame. 

```python
df_freq = pd.DataFrame(lista).T
df_freq.head()
```
